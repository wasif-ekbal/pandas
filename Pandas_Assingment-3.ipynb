## Q3. Make a function to find top bowler(in terms of most wickets) against any team. Take input as team.

```python
def check(team):
  out = ['caught', 'bowled', 'lbw', 'caught and bowled','stumped']
  team=changer[team.upper()]
  result = delivery[(delivery['batting_team']==team)&(delivery['dismissal_kind'].isin(out))]['bowler'].value_counts().head(1)
  print(f'top 5 bowlers taken wickets against {team} :- \n{result}')
team = input('Enter the name of the team : ')
check(team)
```

## Q4. Write a function which will take team name as input and return it's record of no. occurance in finals and no. of times won.  [Difficult]

ex.-- if team is CSK-- 
CSK in finals-

Played- 8
Won- 3

If any team which has not played in finals like Gujrat Lions

Gujrat Lions in finals-

Played-0


```python
def final_records(team):
    team=changer[team.upper()]
    finals = matches.drop_duplicates('season',keep='last')
    try:
        x = finals[finals['team1']==team]['team1'].value_counts()[0]
    except:
        x=0
    try:
        y = finals[finals['team2']==team]['team2'].value_counts()[0]
    except:
        y=0
    try:
        wins = finals[finals['winner']==team]['winner'].value_counts()[0]
    except:
        wins=0
    total=x+y
    print(f'played in final : {total}\nWon in final : {wins}')

team = input('Name of team : ')
final_records(team)
```

## Q5. Write a function which will take a input as batsman and in output it gives total runs scored against each team. [Medium]




```python
def total_run(batsman_name):
  final=delivery[delivery['batsman']==batsman_name][['bowling_team','batsman','total_runs']]
  print(f'Following is the list of runs made by {batsman_name}, against each team :- ')
  print(final['bowling_team'].value_counts())

x = input('Enter the player\'s name : ')
total_run(x)
```
